# Análisis exploratorio completo

Vamos a analizar dos tipos de vino: el vino rojo y el vino blanco. 

Utilizamos la librería de PanDas para abrir los datasets.

````
import pandas as pd
````

````
url_wine_red = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/winequality-red.csv'
url_wine_white = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/winequality-white.csv'

red = pd.read_csv(url_wine_red, delimiter=';')
white = pd.read_csv(url_wine_white, delimiter=';')

red
````

````
white
````

Como tenemos dos datasets diferentes y queremos unirlos, creamos una columna de categoría para poder identificar a qué data set corresponde una fila posterior a la unión.

````
red['category'] = 'red'
white['category'] = 'white'

# Uniendo ambos datasets
total_wine = red.append(white, ignore_index=True)
total_wine
````

````
total_wine.dtypes
````

Nos podemos dar una idea rápida de cómo estan distribuidos nuestros datos con la función ```describe```

````
total_wine.describe()
````

Empezamos a graficar para ver cuál variable se dispara más

````
total_wine.plot()
````

Graficamos la densidad de los vinos, con lo cual podemos distinguir algunos valores atípicos

````
total_wine['density'].plot()
````

**¿Cuál es nuestro objetivo de nuestro análisis exploratorio?**:
Entender qué factores afectan la calidad de los vinos

Dicho esto, graficamos la calidad. 

````
import seaborn as sns
````

````
sns.set(rc={'figure.figsize': (14,8)})
sns.countplot(total_wine['quality'])
````

Observando la gráfica, notamos que los vinos pueden tener calidad alta o baja y queremos saber qué factores lo afectan. Para eso, utilizamos un pairplot.

````
sns.pairplot(total_wine)
````

Ahora, analizamos la correlación para entender nuestro negocio.

Viendo específicamente las correlaciones de la calidad con las demás variables, podemos observar que su correlación es baja. Sin embargo, podemos observar una correlación alta entre la densidad y la cantidad de alcohol

````
sns.heatmap(total_wine.corr(), annot=True, fmt='.2f', linewidths=2)
````

Realizamos una gráfica de distribución de los datos del alcohol, porque una hipótesis es de la calidad se encuentra relacionada con la cantidad de alcohol

````
sns.distplot(total_wine['alcohol'])
````

De acuerdo a la gráfica anterior, podemos observar que la asimetría de los datos es positiva por lo que tenemos una mayor cantidad de datos del lado izquierdo de la gráfica.

````
from scipy.stats import skew

skew(total_wine['alcohol'])
````

Realizamos un boxplot del alcohol contra la calidad. Podemos observar que los vinos con una calidad de 5 presenta muchos outliers, por lo que realizar un análisis más a fondo en estos vinos valdría la pena.

````
sns.boxplot(x='quality', y='alcohol', data=total_wine)
````

Para continuar con el análisis, realizamos una división de los vinos de acuerdo al valor de su calidad, esto es:

- calidad baja: 3, 4, 5
- calidad media: 6, 7
- calidad alta: 8 y 9

````
total_wine['quality_label'] = total_wine['quality'].apply(lambda value: ('low' if value <=5 else 'medium') if value <=7 else 'high')
total_wine['quality_label'] = pd.Categorical(total_wine['quality_label'], categories=['low', 'medium', 'high'])
total_wine.dtypes
````

Realizamos los histogramas para saber la distribución de cada variable

````
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````
fig = total_wine.hist(bins=15, color='blue', edgecolor='darkmagenta', linewidth=1.0, xlabelsize=10, ylabelsize=10, xrot=45, yrot=0, figsize=(10,9), grid=False)
plt.tight_layout(rect=(0, 0, 1.5, 1.5))
````

Utilicemos un modelo de regresión para saber cómo se comporta la calidad (con las nuevas categorías ya hechas) y poder realizar una predicción.

````
## Analisis de regresion from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split,cross_validate
from sklearn.preprocessing import MinMaxScaler,StandardScaler,LabelEncoder
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
````

````
label_quality = LabelEncoder()
total_wine['quality_label'] = label_quality.fit_transform(total_wine['quality_label'])
total_wine
````

````
total_wine.columns
````

````
training_set_regresor=total_wine.drop(['category'],axis=1)
````

Realizamos una división, una para el entrenamiento y otra para el testing

````
x_train,x_test,y_train,y_test=train_test_split(training_set_regresor,total_wine['quality_label'],test_size=0.30,random_state=42)
````

````
x_train
````

````
model_names=['KNearestNeighbors']
````

````
acc=[]
eval_acc={}
classification_model=KNeighborsClassifier()
classification_model.fit(x_train,y_train)
pred=classification_model.predict(x_test)
acc.append(accuracy_score(pred,y_test))
eval_acc={'Modelling Algorithm':model_names,'Accuracy':acc} 

eval_acc
````
